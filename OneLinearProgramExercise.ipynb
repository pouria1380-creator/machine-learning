```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df  = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='navy')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
print('coef',reg.coef_)
print('intercept',reg.intercept_)
print('in 2020 perdiction', reg.predict([[2020]]))
828.46507522*2020-1632210.7578554577
```

```python
print(reg.predict([[2030]]))
828.46507522*2030-1632210.7578554577
print(reg.predict([[2045]]))
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='navy')
plt.plot(df.year,reg.predict(df[['year']]),color='green')
```

```python
perdict_data = {'year':[2025,2026]}
perdict_data_df = pd.DataFrame(data=perdict_data)
perdict_data_df
```

```python
perdict_p = reg.predict(perdict_data_df)
```

```python
perdict_data_df['price']= perdict_p
print(perdict_data_df)
print('reg_cof',reg.coef_,'reg_intercept',reg.intercept_)
print('for 2025 price is: ',828.46507522*2025-1632210.7578554575,'for 2026 price is: ',828.46507522*2026-1632210.7578554575)
perdict_data_df.to_csv('example_perdict.csv',index=False)
```
